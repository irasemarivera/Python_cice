```python
def f1(x):
    y = sin(x)
    return y

def f2(x):
    y = sin(x) + sin(5.0*x)
    return y

def f3(x):
    y = sin(x) + exp(-x/10.)
    return y

x = linspace(0, 10*pi, 800)

p1, p2, p3 = plot(x, f1(x), x, f2(x), x, f3(x))

legend(('Función 1', 'Función 2', 'Función 3'), prop = {'size': 10}, loc = 'upper right')
```
