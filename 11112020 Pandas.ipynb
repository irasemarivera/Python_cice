```python
# Correlaciones entre dos variables.
# Una correlación es una relación lineal entre dos variables cuantitativas se representa como el cociente entre la covarianza
# de dos variables y el producto entre sus desviaciones estándar. Puede tomar valores entre -1 y 1, 
# cuanto más cerca de estos extremos se encuentre la relación estamos hablando de correlaciones fuertes,
# también tenemos dependiendo de esa distancia moderadas, débil y muy débil.
import pandas as pd
import numpy as np

df = pd.read_csv("data/base_datos_2008.csv", nrows = 1000000)
```

Para más información, podéis consultar la página:

# Test de Chi-Cuadrado

```python
# El test chi-cuadrado permite comparar y hacer un estudio entre el valor observado y esperado en un conjunto de datos 
# y comprobar si existe dependencia entre variables.

import pandas as pd
import numpy as np

df = pd.read_csv("data/base_datos_2008.csv")

# Sumas diferencias al cuadrado entre nuestros valores observados y los esperados...
```

$$X^2 = \sum{(observado_{i} - esperado_{i})^2}\div(esperado_{i})$$

https://es.wikihow.com/calcular-el-p-valor

```python
# Al fijar una semilla, cuando quiera replicar el código obtendremos los mismos resultados.
np.random.seed(0)

# Seleccionamos los vuelos con origen en las tres ubicaciones seleccionadas.
df = df[df["Origin"].isin(["HOU", "ATL", "IND"])]

# Una vez aplicado el filtro, reordenamos la BD y seleccionamos las primeras 10.000 filas para realizar un test
df = df.sample(frac = 1)
df = df[0:10000]
```

```python
# Creamos una nueva variable BigDelay que tendrá el valor True cuando el retraso de llegada sea > 30 minutos, falso en caso 
# contrario.
df["BigDelay"] = df["ArrDelay"] > 30
observados = pd.crosstab(index = df["BigDelay"], columns = df["Origin"], margins = True) 

# margins = True nos da los totales a la derecha y abajo.
```

```python
# Hemos generado una tabla de contingencia.
# 6927 son los vuelos de Atlanta que se han retrasado menos de 30 minutos.
observados
```

```python
# A continuación importamos el paquete que va a realizar el test
from scipy.stats import chi2_contingency
```

```python
test = chi2_contingency(observados)
```

```python
# El primer dato que nos devuelve es el estadístico (suma de las diferencias al cuadrado)
# El segundo es el p-valor, el p-valor 
# Finalmente la tabla de valores esperados.

test
```

```python
# Convertimos la tabla de los valores esperados en un Dataframe y la visualizamos
esperados = pd.DataFrame (test[3])
```

```python
esperados
```

```python
esperados_rel = round(esperados.apply(lambda r: r/len(df) * 100, axis = 1), 2)
observados_rel = round(observados.apply(lambda r: r/len(df) * 100, axis = 1), 2)
```

```python
observados_rel
```

```python
esperados_rel
```

```python
# Test chi-cuadrado nos permite comparar los elementos esperados vs observados, para poder generar un estadístico
# Si el p-valor < 0,05 hay diferencias significativas. Hay relación entre variables.
# Si el p-valor > 0,05 no hay diferencias significativas. No hay relación entre variables.

test[1]
```

```python
# No podemos afirmar en este caso que haya relación entre variables.
```

# Análisis de datos externos

```python
import pandas as pd
import numpy as np

df = pd.read_csv("../../data/Cap 8/base_datos_2008.csv", nrows = 100000)
# Miraremos cómo detectar aquellos datos que toman valores extremadamente grandes o pequeños (outlier)
```

```python
# Extraemos la columna ArrDelay y eliminamos los datos faltantes de esta columna
x = df["ArrDelay"].dropna()
```

```python
# Seleccionamos los cuartiles
Q1 = np.percentile(x, 25)
Q3 = np.percentile(x, 75)
# Calculamos el rango intercuartílico, diferencia entre el tercer cuartil y el primer cuartil
# La mediana es el Q2.
rangointer = Q3 - Q1
```

```python
umbralsuperior = Q3 + 1.5 * rangointer
umbralinferior = Q1 - 1.5 * rangointer
```

```python
umbralsuperior
```

```python
umbralinferior
```

```python
# Los casos que están por encima del umbralsuperior, están alrededor de... 8% por encima.
np.mean(x > umbralsuperior)
```

```python
# En este segundo caso, los casos por debajo del umbral inferior están alrededor de... 0,2% por debajo.
np.mean(x < umbralinferior)
```

```python
# Por tanto, las medidas no están simétricas. Vamos a ver cómo podemos estudiar varias variables a la vez.
from sklearn.covariance import EllipticEnvelope
```

```python
# Establecemos un modelo que nos va a seleccionar el 1% de los datos que están más alejados de nuestros datos más centrados.
outliers = EllipticEnvelope(contamination = .01)
```

```python
# Definimos una lista de nuestra BD, todas cuantitativas y asumimos que su distribución es Gaussiana.
var_list = ["DepDelay", "TaxiIn", "TaxiOut", "CarrierDelay", "WeatherDelay", "NASDelay", "SecurityDelay", "LateAircraftDelay"]
```

```python
# Definimos de nuevo la variable x con los valores que nos interesan (lista)
x = np.array(df.loc[:, var_list].dropna())
```

```python
# Pasamos a entrenar nuestro modelo.
outliers.fit(x)
```

```python
# Predicción de los valores que forman parte del 1% que queremos detectar.
pred = outliers.predict(x)
```

```python
pred
```

```python
# Nos interesa detectar los -1
elips_outliers = np.where(pred == -1)[0]
```

```python
# Lista de todos los index de aquellos casos que están muy alejados de los valores centrales de las variables estudiadas.
# Sirven para observar en nuestra BD aquellos valores que sean extremos y que requieran de nuestra atención.
elips_outliers
```

```python
import pandas as pd
import numpy as np

# Utilizamos la biblioteca joblib para ejecutar procesos en paralelo.
from joblib import Parallel, delayed

df = pd.read_csv("../../data/Cap 8/base_datos_2008.csv", nrows = 100000)
```

```python
df_sub = df[["CarrierDelay", "WeatherDelay", "NASDelay", "SecurityDelay", "LateAircraftDelay"]]
df_sub.head(10)
```

```python
def retraso_maximo (fila): 
    if not np.isnan(fila).any():
        names = ["CarrierDelay", "WeatherDelay", "NASDelay", "SecurityDelay", "LateAircraftDelay"]
        return names[fila.index(max(fila))]
    else:
        return "None"
```

```python
results = []
for fila in df_sub.values.tolist():
    results.append(retraso_maximo(fila))
```

```python
results
```

```python
# Implementamos la función anterior, paralelizando el proceso.
result = Parallel(n_jobs = 2, backend = "multiprocessing") (map(delayed(retraso_maximo), df_sub.values.tolist()))
```

# Paralelizar bucles

```python
import pandas as pd
import numpy as np
from joblib import Parallel, delayed

df = pd.read_csv("../../data/Cap 8/base_datos_2008.csv", nrows = 100000)
```

```python
df_sub = df[["CarrierDelay", "WeatherDelay", "NASDelay", "SecurityDelay", "LateAircraftDelay"]]
df_sub.head(10)
```

```python
# Vamos a trabajar con la estructura creada anteriormente.
# Vamos a buscar el retraso máximo dentro de las filas y campos anteriores.
def retraso_maximo(fila):
    if not np.isnan(fila).any():
        names = ["CarrierDelay", "WeatherDelay", "NASDelay", "SecurityDelay", "LateAircraftDelay"]
        return names[fila.index(max(fila))]
    else:
        return "None"
```

```python
results = []
for fila in df_sub.values.tolist():
    results.append(retraso_maximo(fila))
```

```python
results
```

```python
# Vamos a implementar esto pero paralelizando el proceso.
result = Parallel(n_jobs = 1, backend = "multiprocessing") (
    map (delayed(retraso_maximo), df_sub.values.tolist()))
```
