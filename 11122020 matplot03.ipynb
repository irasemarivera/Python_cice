```python
# Construcci√≥n de histogramas.

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("data/base_datos_2008.csv")
```

```python
# Eliminamos los datos que faltan de nuestra BD, y un subconjunto de las variables que nos interesa estudiar.
# Con inplace = True indicamos que no necesitamos devolver el resultado en otra variable.

df.dropna(inplace = True, subset = ["ArrDelay", "DepDelay", "Distance"])

sns.distplot(df["Distance"])
```

```python
# Con kde = False suprimo la aproximaci√≥n a curva que hace el sistema

sns.distplot(df["Distance"], kde = False)
```

```python
# En este caso puedo ajustar el n√∫mero de bins que tiene nuestro histograma
sns.distplot(df["Distance"], kde = False, bins = 100)
```

```python
# Para comparar dos distribuciones

sns.kdeplot(df["ArrDelay"])
sns.kdeplot(df["DepDelay"])

# Acotamos los ejes ya que hay bastantes outliers
plt.xlim ([-300, 300])
```

```python
# Diagramas m√°s b√°sicos, indicando los vuelos que hayan salido de los distintos puntos.
# Vamos a coger solo 500 datos.

df2 = df[df["Origin"].isin(["ATL", "HOU", "IND"])].sample(frac = 1).head(500)
```

```python
# Visualizamos el Origen en el eje Y y el Retardo de salida en el Eje X

sns.boxplot(x="DepDelay", y = "Origin", data = df2)
```

```python
sns.boxplot(x="DepDelay", y = "Origin", data = df2)
# Si queremos ajustar las cajas, podemos hacer
plt.xlim ([-20, 150])
```
