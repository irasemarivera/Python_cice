22. Obtener las curvas de desviación entre el tiempo programado de un vuelo y el tiempo real del mismo, el tiempo programado de salida y la hora de salida real, y la hora programada de llegada y la hora real. Seleccionar para ello el subconjunto de datos adecuado, agrupando por ejemplo por Origen

```python
import random
import numpy as np
import seaborn as sns
import datetime

df = df.head(500000)
dates = []
origins = np.unique(df.Origin)

#Ponemos la fecha
for i in df.index:
    dates.append(datetime.datetime(year = df.loc[i, "Year"], month = df.loc[i,"Month"], day = df.loc[i,"DayofMonth"]))
df["Date"] = dates
```

```python
origin = random.sample(list(origins), 1)
df2 = df[df["Origin"] == origin[0]]
t_df = df2.groupby(by=["Date"])["CRSElapsedTime", "ActualElapsedTime", "CRSDepTime", "DepTime", "CRSArrTime", "ArrTime"].mean()
ax = sns.lineplot(data = t_df, hue = "Origin")
ax.set(title="Curvas de desviación en {}".format(origin))
```
